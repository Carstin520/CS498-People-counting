```python
import scipy.io as sio
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import numpy as np
```

```python
#load the data
data_mat = sio.loadmat('adc_data.mat')
```

```python
nx = data_mat['nLanes'][0]
peaks, _ = find_peaks(nx, distance=150,prominence=(None,5))
num = peaks.shape[0]
print("number of people:", num)
plt.plot(nx)
plt.plot(peaks, nx[peaks], "x")
plt.plot(np.zeros_like(nx), "--", color="gray")
plt.show()
```
